# Prepare the data

## Import and concatanet all files with the "in_progress" tickets for each iteration

```python
all_in_progress = glob.glob("DataSets/" + "*.csv")
all_in_progress_df = pd.concat(map(lambda filename: pd.read_csv(filename).assign(source=filename), all_in_progress))

all_in_progress_df
```

## Search for the partners tickets
Update "Custom field (Type of Request)" if one of the labels is "partner"

```python
label_cols = [col for col in all_in_progress_df.columns if 'label' in col.lower()]
print(label_cols)
```

```python
def update_column_value(row):
    label_columns = [col for col in all_in_progress_df.columns if 'label' in col.lower()]
    column_to_update = "Custom field (Type of Request)"
    
    for column in label_columns:
        if 'partner' in str(row[column]).lower():
            row[column_to_update] = 'SR - Partner'
    return row[column_to_update]
            
            
all_in_progress_partners = all_in_progress_df
all_in_progress_partners[ "Custom field (Type of Request)"] = all_in_progress_df.apply(update_column_value, axis=1)
all_in_progress_partners       
```

```python
all_in_progress_df['source'].unique()
```

## Filter the columns

Keep only colmns necessary for further analysis

```python
all_in_progress_df_filered = all_in_progress_df[['Issue key', 'Issue id', 'Status', 'Reporter', 'Created', 'Due Date', 'Resolved', 'Custom field (Customer/s)','Custom field (Type of Request)', 'source']]
all_in_progress_df_filered.rename({"Custom field (Customer/s)": "Customer", "Custom field (Type of Request)":"Type of Request"},axis=1, inplace=True)
all_in_progress_df_filered
```

## Add additional column "PI" base on the data srouce 

```python
def find_iteration(source_string):
    iterations = ['PI23.3-1','PI23.3-2','PI6 IP', 'PI6.1','PI6.2', 'PI6.3','PI6.4', 'PI7.1', 'PI7.2', 'PI7.3', 'PI7 IP']
    for iteration in iterations:
        if iteration in source_string:
            return iteration
    return "No info"
    
    
all_in_progress_df_iteration = all_in_progress_df_filered
all_in_progress_df_iteration["Iteration"] = all_in_progress_df_iteration['source'].apply(find_iteration)

all_in_progress_df_iteration   
```

```python
all_in_progress_df_final = all_in_progress_df_iteration
file_path = Path('DataSets/AllCombined/all_in_progress.csv')
all_in_progress_df_final .to_csv(path_or_buf=file_path)
```
