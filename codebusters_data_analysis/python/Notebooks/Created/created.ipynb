# Prepare the data

## Import and concatanet all files with the "created" tickets for each iteration

```python
all_created = glob.glob("../../DataSets/Created/" + "*.csv")
all_created_df = pd.concat(map(lambda filename: pd.read_csv(filename).assign(source=filename), all_created))

all_created_df
```

```python
def update_column_value(row, label, column_to_update):
    label_columns = [col for col in all_created_df.columns if 'label' in col.lower()]
    column_to_update = "Custom field (Type of Request)"
    
    for column in label_columns:
        if str(label) in str(row[column]).lower():
            row[column_to_update] = 'SR - Partner'
    return row[column_to_update]


def find_escalatetions(row):
    label_columns = [col for col in all_created_df.columns if 'label' in col.lower()]
    isEscalated = False
    for column in label_columns:
        if 'escalated' in str(row[column]).lower():
            isEscalated = True    
    return isEscalated
```

## Search for the additional labels
Update "Custom field (Type of Request)" if `partner` or `Partners` label is found

```python
label_cols = [col for col in all_created_df.columns if 'label' in col.lower()]
print(label_cols)
```

```python
all_created_partners = all_created_df
all_created_partners["Custom field (Type of Request)"] = all_created_df.apply(lambda row: update_column_value(row, 'partner', 'Custom field (Type of Request)'), axis=1)
all_created_partners
```

## Look for the tickets with `Escalations` label

```python
all_created_escalated_match = all_created_df
all_created_escalated_match["Is Escalated"] = all_created_df.apply(lambda row: find_escalatetions(row), axis=1)
all_created_escalated_match
```

```python
all_created_df['source'].unique()
```

## Filter the columns

Keep only colmns necessary for further analysis

```python
all_created_df_filered = all_created_escalated_match[['Issue key', 'Issue id', 'Status', 'Reporter', 'Created', 'Due Date', 'Resolved', 'Custom field (Customer/s)','Custom field (Type of Request)', 'source','Is Escalated']]
all_created_df_filered.rename({"Custom field (Customer/s)": "Customer", "Custom field (Type of Request)":"Type of Request"},axis=1, inplace=True)
all_created_df_filered
```

## Add additional column "PI" base on the data srouce 

```python
def find_iteration(source_string):
    iterations = ['PI23.3-1','PI23.3-2', 'PI23.3-3','PI6 IP', 'PI6.1','PI6.2', 'PI6.3','PI6.4', 'PI7.1', 'PI7.2', 'PI7.3', 'PI7 IP']
    for iteration in iterations:
        if iteration in source_string:
            return iteration
    return "No info"
    
    
all_created_df_iteration = all_created_df_filered
all_created_df_iteration["Iteration"] = all_created_df_iteration['source'].apply(find_iteration)

all_created_df_iteration   
```
